## 2) Test Results
Below is an analysis of the results of the Gapminder Test 2017. We start with generalities and then go deeper into each question in particular. 

```python
# Importing Data
rs_df = pd.read_excel("gms17-results.xlsx", sheet_name="result-scores", skiprows=21, nrows=13, usecols="B:C")
rs_df
```

#### 2.1) Number of correct answers vs share of respondents

```python
# Plotting Data
fig, ax1 = plt.subplots()

# Bar Chart on axis1
ax1.bar(rs_df["Number of correct answers"], rs_df["Share of respondents"])

# Cumulative Sum on axis2
ax2 = ax1.twinx()
cumsum = rs_df["Share of respondents"].cumsum()
ax2.plot(rs_df["Number of correct answers"], cumsum, color="red", marker="o")

ax1.set_xlabel("Number of correct answers")
ax1.set_ylabel("Share of respondents")
ax1.set_title("Gapminder Test 2017 Results")
ax2.legend(["Cumulative Share of respondents"], loc="upper left")
ax2.set_ylabel('Cumulative Sum', color='red')

plt.show()
```

#### 2.2) Average number of correct answers

```python
# Average number of correct answers
avg_correct = (rs_df["Number of correct answers"] * rs_df["Share of respondents"]).sum()
float(avg_correct)
```

### 2.2) Country Score Distributions

```python
# Importing Data
csd_df = pd.read_excel("gms17-results.xlsx", sheet_name="country score distributions", skiprows=4, nrows=14, usecols="B:O")
csd_df
```

#### Average Correct Answers

```python
csd_df["Average Correct Answers"] = csd_df.apply(lambda x: (csd_df.columns[1:] * x[1:]).sum(), axis=1)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=7, figsize=(20, 20))
plt.subplots_adjust(hspace=0.5, wspace=0.5) 


for i in range(14):
    ax[i//7, i%7].bar(csd_df.columns[1:-1], csd_df.iloc[i, 1:-1])
    ax[i//7, i%7].set_title(csd_df.iloc[i, 0])
    ax[i//7, i%7].axvline(csd_df["Average Correct Answers"][i], color='red', linestyle='--')
    ax[i//7, i%7].text(3, -0.01, f"Average: {csd_df['Average Correct Answers'][i]:.2f}", horizontalalignment='center', verticalalignment='center', rotation=45, color='white', backgroundcolor='red', fontstyle='oblique')
    
    # Chimps
    ax[i//7, i%7].axvline(4, color='black', linestyle='--')
    ax[i//7, i%7].text(6, 0.15, "Chimps (4)", horizontalalignment='center', verticalalignment='center', rotation=45, color='white', backgroundcolor='black', fontstyle='oblique')
```

#### Better than Random Guessing

```python
csd_df["Better than Random Guessing"] = csd_df.apply(lambda x: x[6:-1].sum(), axis=1)
```

```python
csd_df
```

```python
csd_df.sort_values("Better than Random Guessing", ascending=False)
```

```python
plot_df = csd_df.sort_values("Better than Random Guessing", ascending=False)
avg = plot_df["Better than Random Guessing"].mean()

fig, ax = plt.subplots()

ax.bar(plot_df["Name"], plot_df["Better than Random Guessing"])
ax.set_title("Better than Random Guessing")
ax.set_xlabel("Name")
ax.set_ylabel("Better than Random Guessing")
ax.set_xticklabels(plot_df["Name"], rotation=90)


plt.axhline(avg, color='red', linestyle='--')
plt.text(11, 0.10, f"Average: {avg:.2f}", horizontalalignment='center', verticalalignment='center', rotation=45, color='white', backgroundcolor='red', fontstyle='oblique')
plt.show()
```

### 2.3) Country Percent Per Question

```python
# Importing Data
cppq_df = pd.read_excel("gms17-results.xlsx", sheet_name="country percente per question", skiprows=3, nrows=15, usecols="B:N")
cppq_df
```

```python
cppq_df.loc[0, 'Unnamed: 1'] = "Country"
cppq_df
```

```python
# Clean columns
cppq_df.columns = cppq_df.iloc[0]
cppq_df.drop(0, inplace=True)
cppq_df
```

```python
# Set index to country
cppq_df.set_index("Country", inplace=True)
cppq_df
```

#### Correct Per Question Per Country


```python
cppq_df = cppq_df.astype(float)
```

```python
fig, ax = plt.subplots()
cax = ax.imshow(cppq_df.values, cmap='winter', aspect='auto')

# Add color bar
fig.colorbar(cax, ax=ax, label='Scale')

# Set the x and y axis labels
ax.set_xticks(np.arange(len(cppq_df.columns)))
ax.set_yticks(np.arange(len(cppq_df.index)))

# Set tick labels
ax.set_xticklabels(cppq_df.columns, rotation=90)
ax.set_yticklabels(cppq_df.index)

# Title and labels
plt.title("% Correct Per Question Per Country")
plt.xlabel("Question")
plt.ylabel("Countries")

# Loop over data dimensions and create text annotations.
for i in range(len(cppq_df.index)):
    for j in range(len(cppq_df.columns)):
        text = ax.text(j, i, str(int(cppq_df.values[i, j] * 100)) + "%",
                       ha="center", va="center", color="w")

# Display the plot
plt.show()
```

#### Average Correct Per Question

```python
# Average per question
cppq_acpq_df = cppq_df.mean(axis=0)
cppq_acpq_df
```

```python
fig, ax = plt.subplots()

avg = cppq_acpq_df.mean()

ax.barh(cppq_acpq_df.index, cppq_acpq_df)
ax.set_title("Average Correct Per Question")
ax.set_xlabel("% Correct Responses")
ax.set_ylabel("Question")

# Chimps
plt.axvline(avg, color='red', linestyle='--')
plt.text(avg, 0.33, f"Average: {avg:.2f}", horizontalalignment='center', verticalalignment='center', rotation=45, color='white', backgroundcolor='red', fontstyle='oblique')

# Chimps
plt.axvline(0.33, color='black', linestyle='--')
plt.text(0.33, 0.33, f"Chimps: 0.33", horizontalalignment='center', verticalalignment='center', rotation=45, color='white', backgroundcolor='black', fontstyle='oblique')

plt.show()
```
